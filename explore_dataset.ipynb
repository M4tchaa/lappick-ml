# Ekstrak Informasi

##### 1. CPU

```python
import re
import pandas as pd

def extract_cpu_info(row):
    """
    Mengekstrak informasi CPU dari string 'Laptop' berdasarkan 'GPU' dan 'CPU' kolom.
    Mengembalikan string CPU yang diekstrak atau None jika tidak ditemukan.
    """
    cpu_value = None # Inisialisasi default

    # Tentukan pola utama berdasarkan GPU
    if row['GPU'] == 'Integrated':
        main_pattern = r'(.+)\/(?:.+)\/(?:.+)\/(?:.+)'
    else:
        main_pattern = r'(.+)\/(?:.+)\/(?:.+)\/(?:.+)\/(?:.+)'

    main_match = re.search(main_pattern, str(row['Laptop']))

    if main_match:
        first_part_of_laptop_string = main_match.group(1)
        
        # Cari CPU di bagian pertama string laptop
        # Gunakan re.escape untuk menangani karakter khusus di df['CPU']
        cpu_regex = rf'{re.escape(str(row["CPU"]))}.*'
        cpu_match = re.search(cpu_regex, first_part_of_laptop_string, re.IGNORECASE)
        
        if cpu_match:
            cpu_matched_string = cpu_match.group(0)
            
            if cpu_matched_string.startswith('Apple'):
                cpu_value = row['CPU'] # Gunakan nilai dari kolom CPU asli
            elif '/' in cpu_matched_string:
                sub_cpu_match = re.search(r'(.+)\/', cpu_matched_string)
                if sub_cpu_match:
                    cpu_value = sub_cpu_match.group(1).strip()
                else:
                    cpu_value = cpu_matched_string.strip() # Fallback jika sub-regex gagal
            else:
                cpu_value = cpu_matched_string.strip()
        else:
            # Jika CPU tidak ditemukan oleh regex kedua, gunakan nilai CPU asli dari kolom
            # atau biarkan None jika Anda ingin hanya nilai yang diekstrak
            cpu_value = row['CPU'] 
            
    return cpu_value

def extract_laptop_name(row):
    """
    Mengekstrak nama laptop dari string 'Laptop' berdasarkan 'GPU' dan 'CPU' kolom.
    Mengembalikan string nama laptop yang diekstrak atau None jika tidak ditemukan.
    """
    laptop_name_value = None # Inisialisasi default

    # Tentukan pola utama berdasarkan GPU
    if row['GPU'] == 'Integrated':
        main_pattern = r'(.+)\/(?:.+)\/(?:.+)\/(?:.+)'
    else:
        main_pattern = r'(.+)\/(?:.+)\/(?:.+)\/(?:.+)\/(?:.+)'

    main_match = re.search(main_pattern, str(row['Laptop']))

    if main_match:
        first_part_of_laptop_string = main_match.group(1)
        
        # Cari nama laptop (bagian sebelum CPU) di bagian pertama string laptop
        # Gunakan re.escape untuk menangani karakter khusus di df['CPU']
        name_regex = rf'(.+)\s*{re.escape(str(row["CPU"]))}'
        name_match = re.search(name_regex, first_part_of_laptop_string, re.IGNORECASE)
        
        if name_match:
            laptop_name_value = name_match.group(1).strip()
        else:
            # Jika nama laptop tidak ditemukan, bisa diasumsikan seluruh 'first_part_of_laptop_string' adalah nama
            # atau biarkan None jika Anda ingin lebih ketat
            laptop_name_value = first_part_of_laptop_string.strip()

    return laptop_name_value
```

```python
df4 = df3.copy()

# Terapkan fungsi untuk CPU
df4['Laptop_CPU'] = df4.apply(extract_cpu_info, axis=1)

# Terapkan fungsi untuk Nama Laptop
df4['Laptop_name'] = df4.apply(extract_laptop_name, axis=1)

df4
```

```python
def benchmark(df):
    cpu_score = None
    gpu_score = None
    if df['CPU'] in cpu_benchmark['CPU Model'].values:
        cpu_score = cpu_benchmark[cpu_benchmark['CPU Model'] == df['CPU']]['CPU Score'].values[0]
    elif df['CPU'].startswith('Intel Evo'):
        cpu_score = cpu_benchmark[cpu_benchmark['CPU Model'].str.contains(df['CPU'].replace('Intel Evo ', 'Intel '), na=False, case=False)]['CPU Score'].values[0]
    else:
        cpu_score = cpu_benchmark[cpu_benchmark['CPU Model'].str.contains(df['CPU'], na=False, case=False)]['CPU Score'].values[0]
    
    if df['GPU'] in gpu_benchmark['GPU Model'].values:
        gpu_score = gpu_benchmark[gpu_benchmark['GPU Model'] == df['GPU']]['GPU Score'].values[0]
    else:
        gpu_score = gpu_benchmark[gpu_benchmark['GPU Model'].str.contains(df['GPU'], na=False, case=False)]['GPU Score'].values[0]

    return pd.Series([cpu_score, gpu_score], index=['CPU Score', 'GPU Score'])

df6[['CPU Score', 'GPU Score']] = df6.apply(benchmark, axis=1)
df6.head()
    
```

```python
# Perluas pemetaan CPU ke iGPU
def map_integrated_gpu_extended(cpu_name):
    cpu_name = cpu_name.lower()

    intel_gpu_map = {
        "i9-13900h": "Intel Iris Xe",
        "i7-13700h": "Intel Iris Xe",
        "i7-1360p": "Intel Iris Xe",
        "i7-1355u": "Intel Iris Xe",
        "i7-1255u": "Intel Iris Xe",
        "i7-1165g7": "Intel Iris Xe",
        "i7-1195g7": "Intel Iris Xe",
        "i5-1340p": "Intel Iris Xe",
        "i5-1335u": "Intel Iris Xe",
        "i5-1235u": "Intel Iris Xe",
        "i5-1135g7": "Intel Iris Xe",
        "i5-1155g7": "Intel Iris Xe",
        "i3-1315u": "Intel UHD Graphics",
        "i3-1215u": "Intel UHD Graphics",
        "i3-1115g4": "Intel UHD Graphics",
        "n100": "Intel UHD Graphics",
        "n200": "Intel UHD Graphics",
        "n4020": "Intel UHD Graphics 600",
        "n4500": "Intel UHD Graphics",
        "n5030": "Intel UHD Graphics 605",
        "n5100": "Intel UHD Graphics",
        "pentium silver n6000": "Intel UHD Graphics",

    }

    amd_gpu_map = {
        "ryzen 9 7940hs": "Radeon 780M",
        "ryzen 7 7840hs": "Radeon 780M",
        "ryzen 7 7735hs": "Radeon 680M",
        "ryzen 7 5700u": "Radeon Vega 8",
        "ryzen 5 5600u": "Ryzen 5 5600U with Radeon Graphics",
        "ryzen 5 5500u": "Ryzen 5 5500U with Radeon Graphics",
        "ryzen 5 7520u": "Radeon 610M",
        "ryzen 5 7530u": "Ryzen 5 7530U with Radeon Graphics",
        "ryzen 3 7320u": "Radeon 610M",
        "ryzen 3 5300u": "Radeon Vega 6",
        "ryzen 3 5425u": "Radeon Vega 6",
    }

    for key, value in intel_gpu_map.items():
        if key in cpu_name:
            return value
    for key, value in amd_gpu_map.items():
        if key in cpu_name:
            return value
    return "Integrated"  # fallback if no match

# Terapkan fungsi pemetaan yang diperluas
df_cleaned_extended = df6.copy()
df_cleaned_extended.loc[df_cleaned_extended["GPU"] == "Integrated", "GPU"] = df_cleaned_extended.loc[df_cleaned_extended["GPU"] == "Integrated", "CPU"].apply(map_integrated_gpu_extended)

# Simpan versi terbaru ke file
output_path_extended = "cleaned_dataset_v3.csv"
df_cleaned_extended.to_csv(output_path_extended, index=False)

output_path_extended
```
