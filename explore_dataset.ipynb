```python
def extract_laptop_info(row):
    # 'row' di sini adalah Series yang merepresentasikan satu baris dari DataFrame
    if pd.isna(row['GPU']): # Memeriksa apakah nilai GPU di baris ini adalah NaN
        # Jika GPU kosong, asumsikan string 'Laptop' punya 4 bagian
        match = re.search(r'(.+)/(.+)/(.+)/(.+)', row['Laptop'])
        if match:
            return match.groups()
        return None
    else:
        # Jika GPU ada, asumsikan string 'Laptop' punya 5 bagian
        match = re.search(r'(.+)/(.+)/(.+)/(.+)/(.+)', row['Laptop'])
        if match:
            return match.groups()
        return None
```

```python
df4 = df3.copy()

# Terapkan fungsi ke setiap baris
df4['Laptop_CPU'], df4['Laptop_CPU'] = df4.apply(extract_laptop_info, axis=1)
df4
```
